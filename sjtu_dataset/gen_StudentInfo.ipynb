```python
import pandas as pd

submit_df = pd.read_csv('./Data_SubmitRecord.csv')
```

```python
submit_df['time'] = pd.to_datetime(submit_df['time'])
```

```python
submit_df.head()
```

```python
# 把2024-02-19 00:00:00 至 2024-06-23 23:59:59 的数据提取出来
submit_df = submit_df[(submit_df['time'] >= '2024-02-19 00:00:00') & (submit_df['time'] <= '2024-06-23 23:59:59')]
# 按照时间排序
submit_df = submit_df.sort_values(by='index', ascending=True)
# 重置索引
submit_df = submit_df.reset_index(drop=True)
submit_df.head()
```

```python
submit_df.to_csv('./Data_SubmitRecord_2024Spring.csv', index=False)
```

```python
student_df = pd.DataFrame(columns=['index','student_ID','sex','age','major'])
stu_ids = submit_df['student_ID'].unique()
stu_ids.sort()

student_df['index'] = range(len(stu_ids))
student_df['student_ID'] = stu_ids
student_df['sex'] = 'Unknown'
student_df['age'] = -1
student_df['major'] = student_df['student_ID'].apply(lambda x: "M" + x[3:6])
```

```python
student_df.head()
```

```python
student_df.to_csv('./Data_StudentInfo.csv',index=False)
```
