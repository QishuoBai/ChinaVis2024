```python
import pandas as pd
import json
from tqdm import tqdm
import random
import math
import numpy as np

titleInfo_df = pd.read_csv("../dataset/Data_TitleInfo.csv")
df = titleInfo_df.sort_values(by=['sub_knowledge'], ascending=True)
mergeData = pd.read_csv('../dataset/Data_MergedNew.csv')
studentinfo_df = pd.read_csv('../dataset/Data_StudentInfo.csv')
```

```python
# 视图中，知识点背景的绘制数据
OutPutArr = []
for k in df['knowledge'].unique().tolist():
    titNum = 0
    OutObj = {"knowledge": k}
    sklist = df[df['knowledge']==k]['sub_knowledge'].unique().tolist()
    
    skarr=[]
    for sk in sklist:
        skobj = {"sk": sk, "titles": []}
        arr = []
        for index,row in df[df['sub_knowledge'] == sk].iterrows():
            tobj = {"tid": row['title_ID'], "fs": int(row['score'])}
            arr.append(tobj)
            arr = sorted(arr, key=lambda x: x['fs'], reverse=True)
        skobj['titles'] = arr
        skobj['titlesNum'] = len(arr)
        titNum += len(arr)
        skarr.append(skobj)
    OutObj["skarr"] = skarr
    OutObj["skNum"] = len(skarr)
    OutObj["titNum"] = titNum
    OutPutArr.append(OutObj)
```

```python
json_str = json.dumps(OutPutArr, ensure_ascii=False)
with open("StuViewData_title.json", "w", encoding="utf-8") as file:
    file.write(json_str)
```

```python
# 每个学生的题目表现



titleList = titleInfo_df['title_ID'].unique().tolist()

outarr = []
for stu in tqdm(studentinfo_df['student_ID']):
    stuobj = {"sid":stu}
    arr = []
    df1 = mergeData[mergeData['student_ID']==stu]
    for title in titleList:
        df2 = df1[df1['title_ID']==title]
        # 计算人题最高分
        maxScore = int(df2['score'].max()) if df2.shape[0]!=0 else 0
        subNum = int(df2.shape[0]) if df2.shape[0]!=0 else 0
        titobj = {"tid":title, "maxscore":maxScore, "subNum":subNum}
        arr.append(titobj)
    stuobj['scores'] = arr
    outarr.append(stuobj)
outarr
```

```python
json_str2 = json.dumps(outarr, ensure_ascii=False)
with open("StuTitlePerfom.json", "w", encoding="utf-8") as file:
    file.write(json_str2)
```
